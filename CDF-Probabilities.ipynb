```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('C:/Users/Dell/Downloads/weight-height male-female - weight-height.csv')
df.head(2)
```

Use the normal cumulative distribution function with the mean and standard deviation of female height to calculate the probability that a female's height is:

```python
female = df.loc[df['Gender']=='Female'].copy()
female_height = female['Height']

mean = np.mean(female_height)
std = np.std(female_height)
```

1. between 55.0 and 56.0 inches

```python
low_end = 55.0
high_end = 56.0

fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=female_height, stat='density',kde=True)
ax.set_title('Distribution of Female Height')
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
ax.axvline(mean,color='k',ls='--', label=f"Mean: {mean:.2f}")
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=female_height, stat='density',kde=True)
ax.set_title('Distribution of Female Height')
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
ax.axvline(std,color='k',ls='--', label=f"Mean: {mean:.2f}")
```

```python
stats.norm.cdf(56, loc=mean,scale=std) - stats.norm.cdf(55,loc=mean, scale=std)
```

The probability of female height between 55 and 56 is: 0.0015

2. is less than 5 feet (60 in)

```python
low_end= female_height.min()
high_end= 60
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=female_height, stat='density',kde=True)
ax.set_title('Distribution of Female Height')
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0);
```

```python
proba = stats.norm.cdf(60, loc=np.mean(female_height), scale= np.std(female_height))
proba
```

The probability of females height less than 5 is 0.084

```python
low_end = mean
high_end = female_height.min()

fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=female_height, stat='density',kde=True)
ax.set_title('Distribution of Female Height')
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
```
